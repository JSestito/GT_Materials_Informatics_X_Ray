```python
#Inputs
fname = 'D:\Libraries\Documents\Materials Informatics\LiCoO2_Al2O3_20160330.hdf5'
datasets = ['S026','S027','S028'] # name of raw dataset
procRefName = 'bkg_corrected' # reference name to append to rawData
```

```python
# open h5file
f = h5.File(fname, mode='r') 
# Always open in read mode 'r'.
# Until you're ready to write into the file. This is especially true if your OS is not Unix-based like Windows.
rawg = f['20160330/Raw']
procg = f['20160330/Process']
meta = f['20160330/Meta']
```

```python
i = 0;
for name in datasets:
    rawDset = rawg[name]
    procDset = f[rawDset.attrs['bkg_corrected']]   

    if i == 0:
        data = rawDset[:,:,:]
        Cdata = procDset[:,:,:]
                    
        twotheta = rawDset.attrs['delta']
        
        i = 1
    else:
        data = np.vstack([data, rawDset[:,:,:]])
        Cdata = np.vstack([Cdata, procDset[:,:,:]])
        
        twotheta = np.hstack([twotheta,rawDset.attrs['delta']])
        
```

```python
##Insert DELETE CODE
delete = [0]

data_new = np.vstack([data[1:25,:,:],data[42:88,:,:]])
Cdata_new = np.vstack([Cdata[1:25,:,:],Cdata[42:88,:,:]])
twotheta_new = np.hstack([twotheta[1:25],twotheta[42:88]])
```

```python
normIntC = np.mean(Cdata_new,axis=(1,2))
normInt= np.mean(data_new,axis=(1,2))
print normInt.shape
    
fig,ax = plt.subplots()
ax.set_title('Mean of all Pixels')
ax.plot(twotheta_new,normInt,marker = 'o', markerfacecolor='r')
ax.plot(twotheta_new,normIntC,marker = 'o', markerfacecolor='g')
ax.set_xlabel('$2\\theta$ (deg.)')
ax.set_ylabel('(Intensity) $ (arb. units)')       

red_patch = mpatches.Patch(color='red', label='Raw Data')
green_patch = mpatches.Patch(color='green', label='Corrected Data')
ax.legend(handles=[red_patch,green_patch], loc=2)
```
